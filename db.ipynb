```python
def display_records_paginated(cursor, page_size=10):
    """
    Отображает записи из курсора SQL постранично, запрашивая у пользователя
    подтверждение для продолжения.

    Args:
        cursor: Объект курсора базы данных, содержащий результаты запроса.
        page_size: Количество записей для отображения на одной странице (по умолчанию 10).
    """
    offset = 0
    while True:
        # Получаем следующие 'page_size' записей
        cursor.execute(f"SELECT * FROM your_table LIMIT {page_size} OFFSET {offset}")
        records = cursor.fetchall()

        if not records:
            print("Больше записей нет.")
            break

        print(f"\n--- Записи {offset + 1} - {offset + len(records)} ---")
        for i, record in enumerate(records):
            print(f"{offset + i + 1}. {record}")

        offset += len(records)

        if len(records) < page_size:
            print("Все записи отображены.")
            break

        while True:
            user_input = input("Показать следующие 10 записей? (да/нет): ").lower().strip()
            if user_input in ("да", "нет"):
                break
            else:
                print("Некорректный ввод. Пожалуйста, введите 'да' или 'нет'.")

        if user_input == "нет":
            print("Программа завершена.")
            break
```
